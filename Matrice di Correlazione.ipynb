## Calcolo Correlazioni tra chiusure indici e features del dataset
##### Inizio con l'esplorare le correlazioni tra le chiusure e le restanti colonne del dataset

```python
# US30

import numpy as np
import pandas as pd
from tabulate import tabulate

df = pd.read_excel("Complessivo.xlsx")

features = ["US30.Close", "US100.Close", "US500.Close", "US30.ATR", "US100.ATR", "US500.ATR", "US30.Volume",
            "US100.Volume", "US500.Volume", "US30.RSI 14", "US100.RSI 14", "US500.RSI 14", "US30.EMA 50",
            "US30.EMA 200", "US100.EMA 50", "US100.EMA 200", "US500.EMA 50", "US500.EMA 200", "US30.VWAP Mensile",
            "US100.VWAP Mensile", "US500.VWAP Mensile", "US30.Delta CloseVWAP", "US100.Delta CloseVWAP",
            "US500.Delta CloseVWAP", "WilliamsVixFix"]

X = df[features].to_numpy()

correlations = np.corrcoef(X)

table_headers = ["Feature", "Correlazione con US30.Close"]
table_data = []

for i in range(1, len(features)):
    table_data.append([features[i], correlations[0, i]])

table_data = sorted(table_data, key=lambda x: x[1], reverse=True)

print("Correlazione tra US30.Close e altre features (ordinato per correlazione decrescente):")
print(tabulate(table_data, headers=table_headers, tablefmt="grid"))
```

## Matrice di Correlazione
##### Procedo con lo sviluppo della matrice di correlazione e la relativa rappresentazione grafica con 2 disposizioni diverse dei dati

```python
import numpy as np
import pandas as pd
from tabulate import tabulate

df = pd.read_excel("Complessivo.xlsx")

features = ["US30.Close", "US100.Close", "US500.Close", "US30.ATR", "US100.ATR", "US500.ATR", "US30.Volume",
            "US100.Volume", "US500.Volume", "US30.RSI 14", "US100.RSI 14", "US500.RSI 14", "US30.EMA 50",
            "US30.EMA 200", "US100.EMA 50", "US100.EMA 200", "US500.EMA 50", "US500.EMA 200", "US30.VWAP Mensile",
            "US100.VWAP Mensile", "US500.VWAP Mensile", "US30.Delta CloseVWAP", "US100.Delta CloseVWAP",
            "US500.Delta CloseVWAP", "WilliamsVixFix"]

correlations = np.corrcoef(df[features].values, rowvar=False)

table_headers = ["Feature", "Correlazione con US30.Close"]
table_data = []

for i, feature in enumerate(features):
    correlations_with_feature = correlations[i]
    for j, corr_value in enumerate(correlations_with_feature):
        if i != j:
            table_data.append([feature, features[j], corr_value])

table_data = sorted(table_data, key=lambda x: abs(x[2]), reverse=True)

print("Correlazione tra ciascuna feature e tutte le altre features (ordinato per valore assoluto della correlazione decrescente):")
print(tabulate(table_data, headers=["Feature", "Altra Feature", "Correlazione"], tablefmt="grid"))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("Complessivo.xlsx")

features = ["US30.Close", "US100.Close", "US500.Close", "US30.ATR", "US100.ATR", "US500.ATR", "US30.Volume",
            "US100.Volume", "US500.Volume", "US30.RSI 14", "US100.RSI 14", "US500.RSI 14", "US30.EMA 50",
            "US30.EMA 200", "US100.EMA 50", "US100.EMA 200", "US500.EMA 50", "US500.EMA 200", "US30.VWAP Mensile",
            "US100.VWAP Mensile", "US500.VWAP Mensile", "US30.Delta CloseVWAP", "US100.Delta CloseVWAP",
            "US500.Delta CloseVWAP", "WilliamsVixFix"]

correlations = np.corrcoef(df[features].values, rowvar=False)

plt.figure(figsize=(12, 10))
sns.set(font_scale=1)
sns.heatmap(correlations, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=features, yticklabels=features, cbar=True)
plt.title("Matrice di Correlazione")
plt.xticks(rotation=90)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("Complessivo.xlsx")

features = ['US30.Close', 'US30.ATR', 'US30.Volume', 'US30.RSI 14', 'US30.EMA 50', 'US30.EMA 200', 'US30.VWAP Mensile',
            'US30.Delta CloseVWAP', 'US100.Close', 'US100.ATR', 'US100.Volume', 'US100.RSI 14', 'US100.EMA 50', 'US100.EMA 200',
            'US100.VWAP Mensile', 'US100.Delta CloseVWAP', 'US500.Close', 'US500.ATR', 'US500.Volume', 'US500.RSI 14',
            'US500.EMA 50', 'US500.EMA 200', 'US500.VWAP Mensile', 'US500.Delta CloseVWAP', 'WilliamsVixFix']

# Ordina il DataFrame in base all'ordine desiderato delle features
df = df[features]

correlations = np.corrcoef(df.values, rowvar=False)

plt.figure(figsize=(12, 12))
sns.set(font_scale=0.7)
sns.heatmap(correlations, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=features, yticklabels=features, cbar=True)
plt.title("Matrice di Correlazione")
plt.xticks(rotation=90)
plt.show()
```
